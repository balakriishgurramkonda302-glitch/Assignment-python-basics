**7. What are the advantages of using generators over regular functions?**

Memory Efficiency: Generators produce items one at a time and only when requested. This means they don't store the entire sequence in memory, making them highly memory-efficient for working with very large or infinite sequences.

Lazy Evaluation: Values are computed on the fly, only when they are needed. This can improve performance by avoiding unnecessary computations.

Iteration: Generators are iterators, meaning you can use them directly in for loops, next() calls, and other constructs that expect an iterable.

Pause and Resume: When a yield statement is encountered, the function's state (including local variables and the instruction pointer) is saved, and the yielded value is returned to the caller. When next() is called again on the generator, the function resumes execution from where it left off.

Simpler Code for Iterators: Writing a generator function with yield is often much simpler and more readable than writing a custom iterator class that implements __iter__() and __next__() methods.

**1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.**

```python
def all_even(num):
  total=0
  for i in num:
    if i%2==0:
      total+=i
  return total

```

```python
all_even([1,2,3,4])
```

**4.Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n < 2 or n > 200:
        return "Number must be between 1 and 200"

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
is_prime(50)
```

**5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

```python
def fibonacci(limit):
    a, b = 0, 1
    count = 0

    while count < limit:
        yield a
        a, b = b, a + b
        count += 1
```

```python
for i in fibonacci(5):
  print(i)
```

**6.Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(max_exp):
    for i in range(max_exp + 1):
        yield 2 ** i
```

```python
for i in powers_of_two(5):
  print(i)
```

**7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

**9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
celsius = [0, 20, 30, 40, 100]

def c_to_f(temp):
    return (temp * 9/5) + 32


fahrenheit = list(map(c_to_f, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

**10.Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def remove_vowel(ch):
    return ch.lower() not in 'aeiou'


text = "Hello World"


result = ''.join(filter(remove_vowel, text))

print("Original:", text)
print("Without vowels:", result)
```
