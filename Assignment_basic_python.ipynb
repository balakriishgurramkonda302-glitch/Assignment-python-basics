**14. Write a program to reverse a string using a while loop.**

```python
text = input("Enter a string: ")

reverse_text = ""
i = len(text) - 1

while i >= 0:
    reverse_text += text[i]
    i -= 1

print("Reversed string:", reverse_text)
```
